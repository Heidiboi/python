## 최대 공약수

#### GCD : Greatest Common Divisor

```python
# 21, 14 => 7
n1 = 21
n2 = 14
for i in range(min(n1, n2), 0, -1):
    if n1 % i == 0 and n2 % i == 0:
        print(f'최대 공약수는 {i}')
        break
```

#### 유클리드 호제법Euclidean agorithm

- x와 y가 있고 x를 y로 나눈 나머지 r이 있을 때
- x와 y의 최대 공약수는 y와 r의 최대 공약수와 같다

```python
x = 16
y = 14
r = x % y
print(f'x: {x}, y: {y}, r: {r}')

x = y
y = r
r = x % y
print(f'x: {x}, y: {y}, r: {r}')

print(f'r의 값이 0이 되었으므로 x 16과 y 14의 최대 공약수는 {y}')
```

```python
def gcd(x, y):
    if y == 0:
        return x
    else:
        return gcd(y, x % y)

x = 16
y = 14

print(f'두 정숫값의 최대 공약수는 {gcd(x, y)}입니다.')
```

```python
def gcd(x, y):
    while True:
        r = x % y
        if r == 0: break
        x, y = y, r
    return y
x = 16
y = 14

print(f'두 정숫값의 최대 공약수는 {gcd(x, y)}입니다.')
```

## 최소 공배수

#### LCD : Least Common Multiple

```python
# 21, 14 => 7
n1 = 21
n2 = 14
for i in range(max(n1, n2), n1 * n2):
    if i % n1 == 0 and i % n2 == 0:
        print(f'최소 공배수는 {i}')
        break
```

#### 유클리드 호제법Euclidean agorithm

- x와 y가 있고 x * y를 최대 공약수로 나눈 몫은 최소 공배수이다

```python
x = 16
y = 14

def lcm(x, y):
    result = (x * y) // gcd(x, y)
    return result

print(f'두 정숫값의 최소 공배수는 {lcm(x, y)}입니다.')
```
