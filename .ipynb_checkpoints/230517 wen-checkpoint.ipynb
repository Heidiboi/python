```python
#그리디에 대해 소나기를통해
import requests,re
res= requests.get('http://ggoreb.com/quiz/소나기.txt')
res.encoding=None
text=res.text
text=text.replace('\n',"")
pattern=re.compile('소녀.+?[.]') #그리디함을 제거하는 방법, 사스가 센세
result=pattern.findall(text)
#print(result)
for i in result:
    print(i)
```

```python
#finditer<< 잘 안씀
# 꼭 써야하는 경우가 있음
# .com과 .net으로 끝나는 이메일 주소 찾기
#finditer -> ()그룹하 한것도 일단 다 들고오는 특성이 있음
import re
pattern = re.compile('\w+@\w+[.](com|net)')
text = '''
주로 사용하는 이메일 주소는 a@ggoreb.com 이고,
가끔 사용하는 이메일 주소는 b@ggoreb.info 입니다.
이번에 새로 만든 c@ggoreb.net도 있습니다.
'''
result = pattern.finditer(text)
print(result) #???
for r in result:
    print(r.group())

    
# sub를 통해 바꾸기
#result = pattern.sub(change, text) #change= 바꿀 문자열, text 기본 문자열
#일부만 sub로 바꾸기
# 남겨두고자하는 부분을 그룹허
import re
pattern = re.compile('\d{6}-\d{7}')
text = '791111-1234567'
change = '*******'
result = pattern.sub(change, text)
print(result)
import re
pattern = re.compile('(\d{6}-)\d{7}')
text = '791111-1234567'
change = '\g<1>*******'
result = pattern.sub(change, text)
print(result)
```

```python
#연습문제, 센세랑
import re
pattern = re.compile('(\d{3})-\d{3,4}-(\d{3})')
text = '010-8478-8181'
change = '\g<1>-****-\g<2>'
res = pattern.sub(change, text)
print(res)

import re
pattern = re.compile('(\w+@)\w+([.]\w+)')
user_list = ['ggoreb@naver.com', 'seorab@gmail.com',
'human@ggoreb.net']
change = '\g<1>*****\g<2>'
for user in user_list:
    res = pattern.sub(change, user)
    print(res)
```

```python
#센세 문제
import requests,re
res=requests.get('http://ggoreb.com/quiz/pattern.html')
text=res.text
#print(text)
idx=text.find('<!--')
idx=text.find('<!--',idx+1)
text2=text[idx:]
#print(text2)
pattern=re.compile("[a-z]+[A-Z]{3}([a-z]{1})[A-Z]{3}[a-z]+")
#센세 풀이
#pattern=re.compile("[^A-Z][A-Z]{3}([a-z]{1})[A-Z]{3}[^A-Z]")
result=pattern.findall(text2)
print(result)
print("".join(result))
print(len(result))

print()


#난제 나중에 알려준다고하심
text2='aAAAzBBBbCCCxDDD'
print(text2)
pattern=re.compile("[^A-Z][A-Z]{3}([a-z])[A-Z]{3}[^A-Z]")
#|[^A-Z][A-Z]{3}([a-z]{1})[A-Z]{3}$
result=pattern.findall(text2)
print(result)
```

```python
class Book:
    #title='' 방법 1
    #price=''
    def setTitle(self,title):
        self.title=title # 방법 2 이걸 더 권장 
        # 우항의 title=매개변수 ,좌항의 self.title=객체가 가진 전역변수
    def setPrice(self,price):
        self.price=price
# 코드 작성
book1 = Book()
book1.setTitle('Python')
book1.setPrice(10000)
print(book1.title, book1.price)
book2 = Book()
book2.setTitle('Java')
book2.setPrice(10000)
print(book2.title, book2.price,type(book2.title))
```

```python
class Student:
    def __init__(self): #초기화 함수, 위에 오류가 일어나는것을 방지하기 위해 미리 생성
        self.num = 1    #생성자
        self.grade = 1
        self.name = 'Anonymous'
    def setNum (self, num):
        self.num = num
    def setGrade (self, grade):
        self.grade = grade
    def setName (self, name):
        self.name = name
s = Student()
print(s.num, s.grade, s.name)
```
