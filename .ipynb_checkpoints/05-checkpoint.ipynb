```python
#함수 사용1
# 절대값 abs()
print(abs(-99999))
# 반복가능한 자료가 모두 참이면 True
print(all([1, 2, 0]))
print(all(['a', 'b', ' ']))
#연습문제1
# 그래프에 이동한 거리 Dx,dy를 구할 수 있는 함수 distance()
def distance(start, end):
    dx=end[0]-start[0]
    dy=end[1]-start[1]
    return abs(dx),abs(dy)
start = (1, 1)
end = (5, 3)
dx, dy = distance(start, end)
print('dx:', dx, 'dy:', dy)
#연습문제2
print('컴파일 언어가 아닌 것은?')
answer1 = input('1. C 2. C# 3. Python 4. swift => ')
print('컴퓨터 장치가 아닌 것은?')
answer2 = input('1. HDD 2. SDD 3. RAM 4. CPU => ')
print('파이썬의 자료형이 아닌 것은?')
answer3 = input('1. str 2. tuple 3. list 4. Map => ')
answer_list = list()
answer_list.append(answer1)
answer_list.append(answer2)
answer_list.append(answer3)
if all(answer_list):
    print('제출 완료')
else:
    print('재입력')
```

```python
#함수 사용2
# iterable자료에서 하나라도 참이면 True 를 주는 any() <-> all()
print(any([1, 2, 0]))
print(any(('a', 'b', '')))
# ord() <-> chd() 아스키 코드 변환
# dir() 객체가 가지고 있는 변수와 함수를 돌려주는 함수
print(dir([1, 2, 3]))
class MyClass:
    age = 10
    name = 'ggoreb'
    def myFunc():
        return age
print(dir(MyClass))
#divmod() 몫,나머지
print(divmod(-10, 2))
#iterable 인데스와 요소를 반환 enumerate)()
for i, e in enumerate({1: 'a', 2: 'b', 3: 'c'}):
    print(i, e)
#연습문제1
a = 123
b = 16
quotient,remaind=divmod(a,b)
print('몫', quotient)
print('나머지', remaind)
#연습문제2
char_list = ['a', 'b', 'c', 'd', 'e', 'f', 'g']
char_dict = {}
# 코드 작성
for i,e in enumerate(char_list):
    char_dict[i+1]=char_list[i]
print(char_dict)
#
```

```python
#센세랑 연습문제
hexa_string1 = '48 45 4C 4C 4F'
hexa_string2 = '47 47 4F 52 45 42'
def show_hex_to_ch(hexa_string):
    a=[chr(int(i,16)) for i in list(hexa_string.split())]
    return print("".join(a))
show_hex_to_ch(hexa_string1)
show_hex_to_ch(hexa_string2)
a=show_hex_to_ch(hexa_string1)
print(a) #이걸 되돌리는 방법?
print(str(a))
print(list(a))
```
