```python
#함수
def total(a, b, c):
    return a + b + c
def average(num, count):
    return num / count
num = total(a=90, b=80, c=70) # 매개변수 지정 후 값 입력
avg = average(count=3, num=num) # ##순서 상관없이 입력 가능
print(num, avg)

print()

def total(*args): # Tuple 형태로 전달 *arguments 들 이라는 의미. 입력을 여러개 할떄
    total = 0
    for n in args:
        total += n
    return total
num = total(1, 2, 3)
print(num)
num = total(1, 2, 3, 4, 5)
print(num)

print()

def func(**kwargs): # Dictionary 형태로 저장
    print(kwargs)
func(a=1, b=2, c=3)

print()

def setInfo(name, age, isPass=True): #기본값 isPass 은 마지막에만 놔야함
    print(name, age, isPass)
setInfo('ggoreb', 20)
setInfo('ggoreb', 20, False)

print()

def funcA():
    return 'a'
retA = funcA() #함수라는 형태도 자료형으로 취급이 가능하다
print(retA)

c=print
c('dd') # 이렇게도 가능함
```

```python
#연습문제
#1
def is_odd(a):
    if a%2==1:
        return '홀'
    else:
        return '짝'
print("#1")
print(is_odd(3))
print(is_odd(4))

print()
#2
def total(a):
    sum=0
    for i in range(1,a+1):
        sum+=i
    return sum
print("#2")
print(total(10))

print()
#3
def calculator(a):
    total=sum(a)
    average=total/len(a)
    return (total,average)  # a,b 와 같이 ,를 쓰면 튜플로 됨, 리스트로 감싸주면 리스트

print("#3",calculator([3, 7, 9, 11]))#returb {"힙": 합,"평균":평균} <<딕셔너리로 표현하기

print()
#4
def sequence(n):
    sequence_list=[]
    for i in range(1,n+1):
        for j in range(1,i+1):
            sequence_list.append(j)
    return sum(sequence_list)
      
print("#4")
print(sequence(3))

print()
#5
def middle(a):
    if len(a)%2==1:
        if len(a)//2:
            return a[len(a)//2:len(a)//2+1]
        else:
            return a
    elif len(a)%2==0:
        if len(a)//2:
            return a[len(a)//2-1:len(a)//2+1]
        else :
            return a
print("#5")
print(middle('python'))
print(middle('d'))
print(middle('functions'))

print()
#6
def WaterMelon(a):
    water='수'
    melon='박'
    answer=[]
    for i in range(1,a+1):
        if i%2==1:
            answer.append(water)
        if i%2==0:
            answer.append(melon)
    return "".join(answer)
print("#6")
print(WaterMelon(4))
    
print()
#7
def n_sum(a):
    total=0
    a=str(a)
    for i in range(1,len(a)+1):
        total+=int(a[i-1])
    return total
print("#7")
print(n_sum(79))
```
