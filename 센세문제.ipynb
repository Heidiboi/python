```python
# 센세랑
#2차원 배열 생성
box=[[0,0,0],[0,0,0],[0,0,0]]
box=[0 for row in range(3)]
box=[ [ 0 for col in range(3)] for row in range(3)]

#2차원 배열 입력 
for row in range(3):
    for col in range(3):
        box[row][col]=col+1+row*3
        
for i in range(9):
    box[i//3][i%3]=i+1

print()
#2차원 배열 출력
for row in range(len(box)):
    for col in box[row]:
        print(col,end=" ")
    print()
```

```python
#센세문제2 내가푼거
n=17
k=4
count=0
while n>1:

    if n==1:
        break
        
    if n%k==0:
        n=int(n/k)
        count+=1
    else:
        n-=1
        count+=1

print(count)
```

```python
#센세풀이
n=26
k=3
count=0
loop=0
while 1:
    loop+=1
    if n%k==0:
        n//=k
    else:
        n-=1
    count+=1
    if n==1:break
print(count,loop)

#센세풀이2 : 루프횟수를 줄이는게 실력!
n=7
k=4
count=0
loop=0
while 1:
    loop+=1
    if n%k==0:
        n//=k
        count+=1
    else:
        나머지=n%k #핵심코드
        count+=나머지
        n-=나머지
    if n==0:
        count-=1
        break
print(count,loop)
```
