#### 1. 04_seoul_기온.csv 파일 불러오기

```python
# open()
# readlines()
# close()
f=open('04_seoul_기온.csv')
data=f.readlines()
f.close()
print(data)
```

#### 2. 날짜와 최저기온 데이터 정리

```python
# date_list와 min_list 생성
# 반복문을 사용하여 각 데이터 저장
date_list=[]
min_list=[]
for d in data:
    date_list.append(d.split(',')[0])
    min_list.append(d.split(',')[-2])
print(date_list)
print(min_list)
a=min_list.pop(0)# 별도로 최저기온, 날짜을 빼주고
min_list=list(map(float,min_list))#실수로 만듬


#센세버전
import matplotlib.pyplot as plt
date_list=[]
min_list=[]
for d in data[1:]: # 사스가센세! 이 방식으로 한글을 뺴고 가능
    date_list.append(d.split(',')[0])
    min_list.append(float(d.split(',')[-2])) #이 방법으로 바로 실수로 가능
print(date_list)
print(min_list)

#

import matplotlib.pyplot as plt
max_list=[]
avg_list=[]
for d in data[1:]: #<< 이방식으로 한글을 뺴고 가능
    avg_list.append(float(d.split(',')[1]))
    max_list.append(float(d.split(',')[-1])) #이 방법으로 바로 실수로 가능
print(max_list)
print(avg_list)


```

#### 3. 시각화 준비 (환경설정)

```python
# matplotlib.pyplot 모듈 가져오기
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action='ignore')
plt.xticks(range(len(min_list)),date_list,rotation=100)
plt.plot(min_list)
plt.plot(max_list)
plt.plot(avg_list)
```
