```python
#오버라이드 실행

class Shape:
    def __init__(self, width, height, color):
        self.width = width
        self.height = height
        self.color = color
    def draw(self):
        print('draw')
    def getArea(self):
        print(self.width * self.height)
class Rectangle(Shape):
    pass

class Triangle(Shape):
    def getArea(self): #override 상속받은 함수를 바꾸는 것
        print(int(self.width * self.height/2))

rect = Rectangle(20, 20, 'blue')
rect.draw()
rect.getArea()

tri = Triangle(20, 20, 'red')
tri.draw()
tri.getArea()
```
